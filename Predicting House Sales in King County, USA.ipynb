### Data Wrangling

It is evident that clarity regarding the DataFrame column names is lacking and the DataFrame also contains junk data. Hence, it requires some pre-processing to clean the data for regression modeling.

### Model Development

We can Fit a linear regression model using the  longitude feature <code>'long'</code> and caculate the R^2.

```python
X = df[['long']]
Y = df['price']
lm = LinearRegression()
lm.fit(X,Y)
lm.score(X, Y)
```

Next, we can fit a linear regression model to predict the <code>'price'</code> using the feature <code>'sqft_living'</code> then calculate the R^2.

```python
X1 = df[['sqft_living']]
Y1 = df['price']
lm1 = LinearRegression()
lm1.fit(X1,Y1)
lm1.score(X1,Y1)
```

We can see that individual features such as <code>'long'</code> and <code>'sqft_living'</code> are not strongly correlated with the <code>'price'</code> (target variable). Hence, simple linear regression would probably not be a good model for the prediction of house price. In this case, we can use **multiple linear regression** using a list of several features as independent variables.

We can fit a multiple linear regression model to predict the <code>'price'</code> using a selected list of features:

```python
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]   
X_new = df[features]
Y_new = df['price']
lm_new = LinearRegression()
lm_new.fit(X_new,Y_new)
```

Then, calculate the R^2.

```python
lm_new.score(X_new, Y_new)
```

Next, we will apply data transformation to scale and standardize the data. 
We will reate a list of tuples, the first element in the tuple contains the name of the estimator:

<code>'scale'</code>

<code>'polynomial'</code>

<code>'model'</code>

The second element in the tuple  contains the model constructor

<code>StandardScaler()</code>

<code>PolynomialFeatures(include_bias=False)</code>

<code>LinearRegression()</code>

```python
Input=[('scale',StandardScaler()),('polynomial', PolynomialFeatures(include_bias=False)),('model',LinearRegression())]
```

Thereafter, we can use the list to create a pipeline object to predict the 'price', fit the object using the features in the list <code>features</code>, and calculate the R^2.

```python
pipe = Pipeline(Input)
X_new = X_new.astype(float)
pipe.fit(X_new, Y_new)
y_pipe = pipe.predict(X_new)
pipe.score(X_new, Y_new)
```

### Model Evaluation and Refinement

```python
#Import necessary modules
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
print("done")
```

We will split the data into training and testing sets:

```python
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]    
X = df[features]
Y = df['price']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=1)


print("number of test samples:", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```

Next, we can reate and fit a Ridge regression object using the training data, set the regularization parameter to 0.1, and calculate the R^2 using the test data.

```python
#Import necessary modules
from sklearn.linear_model import Ridge
```

```python
RidgeModel = Ridge(alpha=0.1)
RidgeModel.fit(x_train, y_train)
RidgeModel.score(x_test, y_test)
```

Finally, we can perform a second order polynomial transform on both the training data and testing data, create and fit a Ridge regression object using the training data, set the regularisation parameter to 0.1, and calculate the R^2 utilising the test data provided.


```python
pr = PolynomialFeatures(degree=2)
x_train_pr = pr.fit_transform(x_train)
x_test_pr = pr.fit_transform(x_test)
RidgeModel2 = Ridge(alpha=0.1)
RidgeModel2.fit(x_train_pr, y_train)
RidgeModel2.score(x_test_pr, y_test)
```

### Conclusion
We can see that the R^2 for the Ridge regression model after the second order polynomial transform is ~0.7, (i.e., close to 1). Hence, the multiple regression model created and tuned using ridge regression can be considered fairly robust for predicting the house price in King County, Seattle.
